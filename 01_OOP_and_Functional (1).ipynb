## 1 · Object-Oriented Programming

### 1.0 Why OOP for EU-SILC?

EU-SILC has a natural **hierarchical structure**:

```
Country (e.g. "LU", "BE", "ES")
  └── Survey wave (year 2004–2023)
        └── Household  ← H-file + D-file
              └── Person ← R-file + P-file
```

Without OOP, our code works with raw dictionaries — hard to validate, easy to misuse:

```python
# Fragile: typos, wrong types, no documentation
hh = {"id": 42, "income": "€51,600", "country": "lu", "size": -1}
```

With OOP we get **encapsulation** (data + behaviour in one place), **validation** (Pydantic catches errors on load), and **reusability** (the same class works for all 26 EU countries).

````python
import pandas as pd

# Choose a country — participants: use YOUR country if it has PUF data!
# Available: AT BE BG CY DE DK EE EL ES FI FR HR HU IE IT LT LU LV MT NL RO SE SI SK
COUNTRY = "LU"   # Luxembourg — change to e.g. "BE", "ES", "HU", "IE", "SE"
YEAR    = 2012

h_file = DATA_DIR / f"{COUNTRY}_PUF_EUSILC" / f"{COUNTRY}_{YEAR}h_EUSILC.csv"
r_file = DATA_DIR / f"{COUNTRY}_PUF_EUSILC" / f"{COUNTRY}_{YEAR}r_EUSILC.csv"

# Load a small sample
df_h = pd.read_csv(h_file, nrows=3)
row  = df_h.iloc[0]          # one household as a pandas Series

# Build a plain dict from the row
raw = {
    "household_id":     int(row["HB030"]),
    "country":          str(row["HB020"]),
    "survey_year":      int(row["HB010"]),
    "disposable_income": float(row["HY020"]),   # net disposable income (€)
    "total_income":     float(row["HY010"]),    # gross total income (€)
}
print("Raw dict:", raw)
print()
print("Problems with raw dicts:")
print("  ❌ No validation: raw['country'] = 'lu' (lowercase) slips through")
print("  ❌ No methods: we can't call raw.equivalised_income()")
print("  ❌ No documentation: what is HY020 vs HY010?")
````

### 1.1 A First Household Class

Let's fix those problems one step at a time. We start with a **plain class**.

````python
class Household:
    """One EU-SILC household from the H (data) file."""

    VALID_COUNTRIES = {
        "AT", "BE", "BG", "CY", "CZ", "DE", "DK", "EE", "EL", "ES",
        "FI", "FR", "HR", "HU", "IE", "IT", "LT", "LU", "LV", "MT",
        "NL", "PL", "PT", "RO", "SE", "SI", "SK",
    }

    def __init__(self, household_id: int, country: str, survey_year: int,
                 disposable_income: float, total_income: float):
        country = country.strip().upper()
        if country not in self.VALID_COUNTRIES:
            raise ValueError(f"Unknown country code: {country!r}")
        if not (2004 <= survey_year <= 2030):
            raise ValueError(f"Survey year out of range: {survey_year}")

        self.household_id       = household_id
        self.country            = country
        self.survey_year        = survey_year
        self.disposable_income  = disposable_income   # HY020
        self.total_income       = total_income          # HY010

    @property
    def tax_burden(self) -> float:
        """Tax and social contributions: gross minus net income.
        https://ec.europa.eu/eurostat/statistics-explained/index.php?title=Glossary:Equivalised_disposable_income
        """
        return self.total_income - self.disposable_income

    @property
    def effective_tax_rate(self) -> float | None:
        """Effective rate as fraction; None when gross income is zero."""
        if self.total_income == 0:
            return None
        return self.tax_burden / self.total_income

    def __repr__(self) -> str:
        return (
            f"Household(id={self.household_id}, country={self.country!r}, "
            f"year={self.survey_year}, income=€{self.disposable_income:,.0f})"
        )


# Instantiate from the real data row
hh = Household(
    household_id    = int(row["HB030"]),
    country         = str(row["HB020"]),
    survey_year     = int(row["HB010"]),
    disposable_income = float(row["HY020"]),
    total_income    = float(row["HY010"]),
)
print(hh)
print(f"  Tax burden      : €{hh.tax_burden:,.0f}")
print(f"  Effective tax   : {hh.effective_tax_rate:.1%}" if hh.effective_tax_rate else "  Effective tax: N/A")

# Validation in action
try:
    bad = Household(1, "xx", 2012, 50_000, 60_000)
except ValueError as e:
    print(f"\n✅ Validation caught: {e}")
````

### 1.2 Dataclasses — Less Boilerplate, Same Power

Writing `__init__`, `__repr__`, and `__eq__` manually is tedious. Python's `dataclasses` module generates all three for free.

```
Plain class             Dataclass
─────────────────       ────────────────────────────
def __init__(self, …)   @dataclass
    self.x = x          class Foo:
                            x: int
                            y: str = "default"
```

````python
from dataclasses import dataclass, field
from typing import Optional


@dataclass
class Person:
    """
    One EU-SILC person (16+) from the P (personal data) file.

    Key variables
    -------------
    person_id           PB030  unique person identifier
    household_id        PB030 // 10  links to Household.household_id
    age                 PB140 (birth year) → derived
    sex                 PB150  1=male  2=female
    education           PE040  ISCED level
    labour_status       PL031  self-defined economic status
    employment_income   PY010G  gross employee cash income
    pension_income      PY100G  gross old-age benefits
    """
    person_id:         int
    survey_year:       int
    country:           str
    age:               int
    sex:               int                  # 1=male, 2=female
    education:         Optional[int] = None # ISCED 0-6
    labour_status:     Optional[int] = None # PL031
    employment_income: float = 0.0          # PY010G (€)
    pension_income:    float = 0.0          # PY100G (€)

    # Derived field — NOT a constructor argument
    household_id: int = field(init=False)

    def __post_init__(self):
        # SILC convention: household ID = person ID // 10
        self.household_id = self.person_id // 10
        self.country = self.country.strip().upper()
        if not (0 <= self.age <= 130):
            raise ValueError(f"Unrealistic age: {self.age}")

    @property
    def is_working_age(self) -> bool:
        """True for persons aged 15–64 (EU labour market definition)."""
        return 15 <= self.age <= 64

    @property
    def total_personal_income(self) -> float:
        """Sum of all personal gross income components."""
        return self.employment_income + self.pension_income

    @property
    def sex_label(self) -> str:
        return {1: "Male", 2: "Female"}.get(self.sex, "Unknown")


# Load a real person from the R register file
df_r = pd.read_csv(r_file, nrows=5)
r_row = df_r.iloc[0]

p = Person(
    person_id    = int(r_row["RB030"]),
    survey_year  = int(r_row["RB010"]),
    country      = str(r_row["RB020"]),
    age          = (YEAR - int(r_row["RB080"])) if pd.notna(r_row["RB080"]) else 0,
    sex          = int(r_row["RB090"]),
)

print(p)
print(f"  Household ID   : {p.household_id}")
print(f"  Sex            : {p.sex_label}")
print(f"  Working age    : {p.is_working_age}")

# Dataclasses give __eq__ for free
p2 = Person(person_id=p.person_id, survey_year=p.survey_year, country=p.country,
            age=p.age, sex=p.sex)
print(f"\nEquality check (same data): {p == p2}")
````

### 1.3 Inheritance — A Common Survey Unit Base

Both `Household` and `Person` share year, country, and validation logic. We extract this into a **base class** so we write it once and inherit everywhere.

````python
from dataclasses import dataclass, field
from typing import ClassVar, Optional


@dataclass
class SurveyUnit:
    """
    Abstract base for all EU-SILC survey units.
    Subclasses: Household, Person.
    """
    survey_year: int
    country:     str

    # Class-level constant — shared across all instances
    VALID_YEARS: ClassVar[range] = range(2004, 2031)

    def __post_init__(self):
        self.country = self.country.strip().upper()
        if self.survey_year not in self.VALID_YEARS:
            raise ValueError(f"survey_year {self.survey_year} not in {self.VALID_YEARS}")

    @property
    def label(self) -> str:
        """Short human-readable label for reports and charts."""
        return f"{self.country} {self.survey_year}"


@dataclass
class HouseholdV2(SurveyUnit):
    """EU-SILC household — H file + D file."""
    household_id:      int
    disposable_income: float    # HY020 net
    total_income:      float    # HY010 gross
    household_size:    int = 1  # HX040
    nuts_region:       str = "" # DB040

    def __post_init__(self):
        super().__post_init__()      # ← runs SurveyUnit validation first
        if self.household_size < 1:
            raise ValueError("household_size must be ≥ 1")

    @property
    def equivalised_income(self) -> float:
        """
        Equivalised disposable income per consumption unit.
        Uses the modified OECD scale:
          1.0  for the first adult
          0.5  for each additional adult
          0.3  for each child (< 14)
        Here we use household_size as a rough proxy; the precise
        calculation requires the full person roster (done in notebook 03).
        """
        # Simplified: assume all members are adults (conservative)
        scale = 1.0 + 0.5 * (self.household_size - 1)
        return self.disposable_income / scale

    @property
    def tax_burden(self) -> float:
        return self.total_income - self.disposable_income


@dataclass
class PersonV2(SurveyUnit):
    """EU-SILC person (16+) — P file."""
    person_id:         int
    age:               int
    sex:               int           # 1=M, 2=F
    education:         Optional[int] = None
    employment_income: float = 0.0
    pension_income:    float = 0.0
    household_id:      int = field(init=False)

    def __post_init__(self):
        super().__post_init__()
        self.household_id = self.person_id // 10


# Quick demo — load a few households from Belgium
be_file = DATA_DIR / "BE_PUF_EUSILC" / "BE_2012h_EUSILC.csv"
df_be   = pd.read_csv(be_file, nrows=5)

households = [
    HouseholdV2(
        survey_year       = int(r["HB010"]),
        country           = str(r["HB020"]),
        household_id      = int(r["HB030"]),
        disposable_income = float(r["HY020"]),
        total_income      = float(r["HY010"]),
        household_size    = int(r["HX040"]) if pd.notna(r["HX040"]) else 1,
    )
    for _, r in df_be.iterrows()
]

print(f"Loaded {len(households)} households from Belgium 2012:")
for hh in households:
    print(f"  {hh.label}  id={hh.household_id:<6}  "
          f"income=€{hh.disposable_income:>10,.0f}  "
          f"equiv=€{hh.equivalised_income:>10,.0f}")
````

### 1.4 Pydantic — Production-Grade Validation

Dataclasses are great for structure, but **Pydantic** gives us:
- Automatic **type coercion** (`"2012"` → `2012`)
- Rich **field validators** with custom logic
- **Serialisation** to JSON / dict with one call
- Clear **error messages** when data is wrong

This matters enormously when loading EU-SILC CSVs, where income fields can be empty strings, `"NA"`, or even negative (self-employed losses are valid in SILC).

````python
from pydantic import BaseModel, Field, field_validator, model_validator
from typing import Optional


class HouseholdModel(BaseModel):
    """
    Pydantic model for one EU-SILC household.
    Validates and coerces data on construction.
    """
    survey_year:       int   = Field(..., ge=2004, le=2030, description="SILC wave year")
    country:           str   = Field(..., min_length=2, max_length=2)
    household_id:      int   = Field(..., gt=0)
    disposable_income: float = Field(..., description="HY020 net disposable income (€)")
    total_income:      float = Field(..., description="HY010 gross total income (€)")
    household_size:    int   = Field(1, ge=1, le=30)
    nuts_region:       Optional[str] = None

    # ── validators ──────────────────────────────────────────────────────────

    @field_validator("country", mode="before")
    @classmethod
    def normalise_country(cls, v: str) -> str:
        return str(v).strip().upper()

    @field_validator("disposable_income", "total_income", mode="before")
    @classmethod
    def coerce_income(cls, v) -> float:
        """Accept numeric strings, empty strings → 0, 'NA' → 0."""
        if v is None or str(v).strip() in ("", "NA", "na"):
            return 0.0
        return float(v)

    @model_validator(mode="after")
    def income_sanity(self) -> "HouseholdModel":
        """Warn (not error) on suspicious income relationships."""
        if self.total_income > 0 and self.disposable_income > self.total_income * 1.5:
            import warnings
            warnings.warn(
                f"HH {self.household_id}: disposable ({self.disposable_income:,.0f}) "
                f"> 150% of gross ({self.total_income:,.0f}) — check data"
            )
        return self

    # ── derived ─────────────────────────────────────────────────────────────

    @property
    def equivalised_income(self) -> float:
        scale = 1.0 + 0.5 * (self.household_size - 1)
        return self.disposable_income / scale

    @property
    def tax_burden(self) -> float:
        return self.total_income - self.disposable_income
````

## 2 · Functional Programming

### 2.0 Why Functional for Statistics?

Statistical workflows are full of **transformations**: filter → aggregate → compute. Functional programming makes these pipelines **composable** and **testable**:

| Concept | Use in SILC processing |
|---------|------------------------|
| **Decorators** | Add timing / logging to Gini, AROP functions without touching their bodies |
| **Generators** | Stream 500k CSV rows one-by-one — no RAM blowup |
| **`functools.partial`** | Fix a country argument, get a reusable single-country function |
| **`functools.reduce`** | Accumulate income sums across country chunks |
| **Higher-order functions** | Pass a poverty threshold as a function argument |

### 2.1 Decorators

A decorator is a function that **wraps another function** to add behaviour — without modifying the original code.

```
@timer
def gini(incomes): ...           # gini = timer(gini)
```

Use `functools.wraps` to preserve the wrapped function's `__name__` and docstring.

````python
import time
import functools
import logging

logging.basicConfig(level=logging.INFO, format="%(levelname)s: %(message)s")


def timer(func):
    """Decorator: print how long a function takes."""
    @functools.wraps(func)
    def wrapper(*args, **kwargs):
        t0 = time.perf_counter()
        result = func(*args, **kwargs)
        elapsed = time.perf_counter() - t0
        print(f"⏱️  {func.__name__}() → {elapsed * 1000:.2f} ms")
        return result
    return wrapper


def log_call(func):
    """Decorator: log every call with argument summary."""
    @functools.wraps(func)
    def wrapper(*args, **kwargs):
        n = len(args[0]) if args and hasattr(args[0], "__len__") else "?"
        logging.info(f"→ {func.__name__}(n={n}, **{list(kwargs.keys())})")
        result = func(*args, **kwargs)
        logging.info(f"← {func.__name__} returned {type(result).__name__}")
        return result
    return wrapper


# Stack multiple decorators (applied bottom-up)
@timer
@log_call
def compute_gini(incomes: list[float]) -> float:
    """
    Gini coefficient — pure Python reference implementation.
    Range: 0 (perfect equality) to 1 (one person has everything).
    We will vectorise this with NumPy in notebook 05.
    """
    n = len(incomes)
    if n == 0:
        return 0.0
    sorted_inc = sorted(incomes)
    cumsum     = 0.0
    total      = sum(sorted_inc)
    if total == 0:
        return 0.0
    for rank, val in enumerate(sorted_inc, start=1):
        cumsum += val * (2 * rank - n - 1)
    return cumsum / (n * total)


# Test on Spain 2012 incomes (positive only)
es_incomes = [hh.disposable_income for hh in valid_hh if hh.disposable_income > 0]
gini_es    = compute_gini(es_incomes)
print(f"\nSpain 2012 Gini (first 500 HH, positive incomes only): {gini_es:.4f}")
print("Note: 0 = perfect equality, 1 = maximum inequality. EU average ≈ 0.30")
````

````python
# Parametrised decorator — useful for configurable behaviour
def retry(max_attempts: int = 3, exceptions: tuple = (Exception,)):
    """Decorator factory: retry a function up to max_attempts times on failure."""
    def decorator(func):
        @functools.wraps(func)
        def wrapper(*args, **kwargs):
            last_exc = None
            for attempt in range(1, max_attempts + 1):
                try:
                    return func(*args, **kwargs)
                except exceptions as e:
                    last_exc = e
                    print(f"  Attempt {attempt}/{max_attempts} failed: {e}")
            raise last_exc
        return wrapper
    return decorator


@retry(max_attempts=3, exceptions=(ValueError,))
def fetch_country_data(country: str) -> pd.DataFrame:
    """Load SILC H-file for 2012 — retries on transient errors."""
    path = DATA_DIR / f"{country}_PUF_EUSILC" / f"{country}_2012h_EUSILC.csv"
    if not path.exists():
        raise ValueError(f"No PUF data for {country!r}")
    return pd.read_csv(path)


try:
    df = fetch_country_data("IE")    # Ireland
    print(f"Ireland 2012: {len(df):,} households loaded")
except ValueError as e:
    print(f"Failed after retries: {e}")

try:
    fetch_country_data("BA")         # Bosnia — not in PUF
except ValueError as e:
    print(f"\nBosnian PUF not in dataset (as expected): {e}")
````

### 2.2 Generators — Memory-Efficient Data Streaming

The full EU-SILC PUF spans 26 countries × 10 years × 4 files ≈ **40 million rows**. Loading everything at once would exhaust RAM.

A **generator** uses `yield` instead of `return` — it produces one item at a time and pauses until the next item is requested. Memory cost = **one row**, not the whole dataset.

````python
import csv

def stream_silc_households(country: str, year: int, data_dir: Path):
    """
    Generator: yield one household dict at a time from the H-file CSV.
    Never loads more than one row into memory.
    """
    path = data_dir / f"{country}_PUF_EUSILC" / f"{country}_{year}h_EUSILC.csv"
    with open(path, newline="", encoding="utf-8") as fh:
        reader = csv.DictReader(fh)
        for row in reader:
            yield row            # pause here; resume when caller asks for next


# Count households without loading the full DataFrame
total   = 0
neg_inc = 0
for hh in stream_silc_households("FR", 2012, DATA_DIR):
    total += 1
    income = float(hh["HY020"]) if hh["HY020"].strip() not in ("", "NA") else 0.0
    if income < 0:
        neg_inc += 1

print(f"France 2012 H-file:")
print(f"  Total households          : {total:,}")
print(f"  Households with neg income: {neg_inc} ({neg_inc/total:.2%})")
print("  (Negative incomes are valid in SILC — self-employment losses)")
````

````python
# Generator pipelines — compose small generators into processing chains

def parse_income(rows):
    """Generator: parse HY020 from each row, skip NA."""
    for i, row in enumerate(rows):
        raw = row.get("HY020", "").strip()
        if raw and raw != "NA":
            if i % 500 == 0:
                print("Parse", i)
            yield float(raw)

def filter_positive(values):
    """Generator: pass through only positive values."""
    for i, v in enumerate(values):
        if i % 500 == 0:
            print(" Filter", i)
        if v > 0:
            yield v

def cap_outliers(values, cap_multiple: float = 10.0):
    """
    Generator: Winsorise extreme incomes.
    Values above cap_multiple × mean are replaced with that cap.
    We compute the mean lazily over a buffer of 1000 rows.
    """
    buffer = []
    for i, v in enumerate(values):
        if i % 500 == 0:
            print("  Cap", i)
        buffer.append(v)
        if len(buffer) >= 1000:
            mean = sum(buffer) / len(buffer)
            cap  = cap_multiple * mean
            for j, x in enumerate(buffer):
                if j < 3:
                    print("   ", x)
                yield min(x, cap)
            buffer = []
    # flush remaining
    if buffer:
        mean = sum(buffer) / len(buffer)
        cap  = cap_multiple * mean
        for j, x in enumerate(buffer):
            if j < 3:
                print("   ", x)
            yield min(x, cap)


# Pipeline: rows → parse → filter → cap
raw_rows = stream_silc_households("SE", 2012, DATA_DIR)   # Sweden
incomes  = cap_outliers(filter_positive(parse_income(raw_rows)))

# Materialise only what we need
inc_list = list(incomes)
median   = sorted(inc_list)[len(inc_list) // 2]
print(f"Sweden 2012 — cleaned income pipeline:")
print(f"  N households  : {len(inc_list):,}")
print(f"  Median income : €{median:,.0f}")
print(f"  AROP threshold: €{0.6 * median:,.0f}  (60% of median)")
````

### 2.3 Higher-Order Functions and `functools`

Functions are **first-class objects** in Python — they can be passed as arguments, returned from other functions, and stored in collections.

`functools` provides battle-tested tools for working with functions:

````python
import functools
import itertools
import operator


# ── functools.partial ────────────────────────────────────────────────────────
# Fix some arguments of a multi-argument function

def arop_rate(incomes: list[float], threshold_pct: float = 0.60) -> float:
    """At-risk-of-poverty rate for a list of equivalised incomes."""
    if not incomes:
        return 0.0
    median    = sorted(incomes)[len(incomes) // 2]
    threshold = threshold_pct * median
    return sum(1 for i in incomes if i < threshold) / len(incomes)

# EU standard: 60% of median. But researchers also use 50%, 70%.
arop_50 = functools.partial(arop_rate, threshold_pct=0.50)
arop_60 = functools.partial(arop_rate, threshold_pct=0.60)   # EU definition
arop_70 = functools.partial(arop_rate, threshold_pct=0.70)

inc_list_sorted = sorted(inc_list)  # Sweden incomes from above

print("Sweden 2012 — AROP at different thresholds:")
for name, fn in [("50%", arop_50), ("60% (EU)", arop_60), ("70%", arop_70)]:
    print(f"  AROP ({name:<8}) = {fn(inc_list_sorted):.1%}")
````

````python
%%writefile silc_toolkit/models.py
"""
Pydantic models for EU-SILC survey units.

EU-SILC file structure
----------------------
D (register)  household level  → DB030 household ID, DB040 NUTS region, DB090 weight
R (register)  person level     → RB030 person ID, links to household
H (data)      household level  → HB030 ID, HY020 net income, HX040 size
P (data)      person 16+       → PB030 ID, PY010G employment income
"""
from __future__ import annotations

import statistics
import warnings
from typing import Optional

from pydantic import BaseModel, Field, field_validator, model_validator


class HouseholdModel(BaseModel):
    """One EU-SILC household (H-file row)."""

    survey_year:       int   = Field(..., ge=2004, le=2030)
    country:           str   = Field(..., min_length=2, max_length=2)
    household_id:      int   = Field(..., gt=0)
    disposable_income: float = Field(..., description="HY020 net disposable income (€)")
    total_income:      float = Field(0.0,  description="HY010 gross total income (€)")
    household_size:    int   = Field(1,    ge=1, le=30,   description="HX040")
    weight:            float = Field(1.0,  gt=0,          description="DB090 design weight")
    nuts_region:       Optional[str] = Field(None,        description="DB040")

    @field_validator("country", mode="before")
    @classmethod
    def normalise_country(cls, v: str) -> str:
        return str(v).strip().upper()

    @field_validator("disposable_income", "total_income", mode="before")
    @classmethod
    def coerce_income(cls, v) -> float:
        if v is None or str(v).strip() in ("", "NA"):
            return 0.0
        return float(v)

    @field_validator("household_size", mode="before")
    @classmethod
    def coerce_size(cls, v) -> int:
        if v is None or str(v).strip() in ("", "NA"):
            return 1
        return max(1, int(float(v)))

    @model_validator(mode="after")
    def income_sanity(self) -> HouseholdModel:
        if self.total_income > 0 and self.disposable_income > self.total_income * 1.5:
            warnings.warn(
                f"HH {self.household_id}: disposable > 150% of gross. Check data."
            )
        return self

    @property
    def equivalised_income(self) -> float:
        """Simplified OECD equivalised income (full roster needed for precision)."""
        scale = 1.0 + 0.5 * (self.household_size - 1)
        return self.disposable_income / scale

    @property
    def tax_burden(self) -> float:
        return self.total_income - self.disposable_income


class PersonModel(BaseModel):
    """One EU-SILC person (P-file row, 16+)."""

    survey_year:       int   = Field(..., ge=2004, le=2030)
    country:           str   = Field(..., min_length=2, max_length=2)
    person_id:         int   = Field(..., gt=0)
    age:               int   = Field(..., ge=0, le=130)
    sex:               int   = Field(..., ge=1, le=2,  description="1=M 2=F")
    education:         Optional[int] = Field(None, description="PE040 ISCED level")
    labour_status:     Optional[int] = Field(None, description="PL031")
    employment_income: float = Field(0.0, description="PY010G gross (€)")
    pension_income:    float = Field(0.0, description="PY100G gross (€)")

    @field_validator("country", mode="before")
    @classmethod
    def normalise_country(cls, v: str) -> str:
        return str(v).strip().upper()

    @field_validator("employment_income", "pension_income", mode="before")
    @classmethod
    def coerce_income(cls, v) -> float:
        if v is None or str(v).strip() in ("", "NA"):
            return 0.0
        return float(v)

    @property
    def household_id(self) -> int:
        return self.person_id // 10

    @property
    def is_working_age(self) -> bool:
        return 15 <= self.age <= 64

    @property
    def sex_label(self) -> str:
        return {1: "Male", 2: "Female"}.get(self.sex, "Unknown")


class SurveyWave(BaseModel):
    """All households for one country × year pair."""

    country:    str
    year:       int
    households: list[HouseholdModel] = []

    @property
    def _positive_incomes(self) -> list[float]:
        return [hh.equivalised_income for hh in self.households
                if hh.equivalised_income > 0]

    @property
    def median_income(self) -> float:
        inc = self._positive_incomes
        return statistics.median(inc) if inc else 0.0

    @property
    def poverty_threshold(self) -> float:
        """EU AROP threshold: 60% of median equivalised income."""
        return 0.6 * self.median_income

    @property
    def arop_rate(self) -> float:
        """Share of households below the poverty threshold."""
        thresh = self.poverty_threshold
        if not self.households or thresh == 0:
            return 0.0
        n_poor = sum(1 for hh in self.households if hh.equivalised_income < thresh)
        return n_poor / len(self.households)

    def summary(self) -> dict:
        return {
            "country":            self.country,
            "year":               self.year,
            "n_households":       len(self.households),
            "median_income":      round(self.median_income, 2),
            "poverty_threshold":  round(self.poverty_threshold, 2),
            "arop_rate":          round(self.arop_rate, 4),
        }
````

````python
%%writefile silc_toolkit/functional.py
"""
Functional programming utilities for SILC data processing.
Decorators, generators, and higher-order helpers.
"""
from __future__ import annotations

import csv
import functools
import logging
import time
from pathlib import Path
from typing import Generator, Iterator

logger = logging.getLogger(__name__)


# ── Decorators ───────────────────────────────────────────────────────────────

def timer(func):
    """Print elapsed time for each call."""
    @functools.wraps(func)
    def wrapper(*args, **kwargs):
        t0 = time.perf_counter()
        result = func(*args, **kwargs)
        ms = (time.perf_counter() - t0) * 1000
        print(f"⏱️  {func.__name__}() → {ms:.2f} ms")
        return result
    return wrapper


def log_call(func):
    """Log INFO messages on entry and exit."""
    @functools.wraps(func)
    def wrapper(*args, **kwargs):
        n = len(args[0]) if args and hasattr(args[0], "__len__") else "?"
        logger.info("→ %s(n=%s)", func.__name__, n)
        result = func(*args, **kwargs)
        logger.info("← %s done", func.__name__)
        return result
    return wrapper


def validate_output(min_val: float, max_val: float):
    """Parametrised decorator: assert return value in [min_val, max_val]."""
    def decorator(func):
        @functools.wraps(func)
        def wrapper(*args, **kwargs):
            result = func(*args, **kwargs)
            if not (min_val <= result <= max_val):
                raise ValueError(
                    f"{func.__name__} returned {result:.6f}, "
                    f"expected [{min_val}, {max_val}]"
                )
            return result
        return wrapper
    return decorator


def retry(max_attempts: int = 3, exceptions: tuple = (Exception,)):
    """Retry a function up to max_attempts times on specified exceptions."""
    def decorator(func):
        @functools.wraps(func)
        def wrapper(*args, **kwargs):
            last_exc = None
            for attempt in range(1, max_attempts + 1):
                try:
                    return func(*args, **kwargs)
                except exceptions as e:
                    last_exc = e
                    logger.warning("Attempt %d/%d failed: %s", attempt, max_attempts, e)
            raise last_exc
        return wrapper
    return decorator


# ── Generators ───────────────────────────────────────────────────────────────

def stream_silc_households(
    country: str, year: int, data_dir: Path
) -> Generator[dict, None, None]:
    """Yield raw H-file rows as dicts without loading the full CSV."""
    path = data_dir / f"{country}_PUF_EUSILC" / f"{country}_{year}h_EUSILC.csv"
    if not path.exists():
        logger.warning("No H-file for %s %d at %s", country, year, path)
        return
    with open(path, newline="", encoding="utf-8") as fh:
        for row in csv.DictReader(fh):
            yield row


def stream_silc_persons(
    country: str, year: int, data_dir: Path
) -> Generator[dict, None, None]:
    """Yield raw P-file rows (persons 16+) as dicts."""
    path = data_dir / f"{country}_PUF_EUSILC" / f"{country}_{year}p_EUSILC.csv"
    if not path.exists():
        logger.warning("No P-file for %s %d at %s", country, year, path)
        return
    with open(path, newline="", encoding="utf-8") as fh:
        for row in csv.DictReader(fh):
            yield row


def stream_all_countries(
    countries: list[str], year: int, data_dir: Path
) -> Iterator[tuple[str, float]]:
    """Yield (country, equivalised_income) for every household across countries."""
    for country in countries:
        for row in stream_silc_households(country, year, data_dir):
            inc_str  = row.get("HY020", "").strip()
            size_str = row.get("HX040", "").strip()
            if not inc_str or inc_str == "NA":
                continue
            income = float(inc_str)
            size   = max(1, int(float(size_str))) if size_str and size_str != "NA" else 1
            scale  = 1.0 + 0.5 * (size - 1)
            yield (country, income / scale)


# ── Higher-order helpers ─────────────────────────────────────────────────────

def arop_rate(incomes: list[float], threshold_pct: float = 0.60) -> float:
    """At-risk-of-poverty rate: share of incomes below threshold_pct * median."""
    if not incomes:
        return 0.0
    median    = sorted(incomes)[len(incomes) // 2]
    threshold = threshold_pct * median
    return sum(1 for i in incomes if i < threshold) / len(incomes)


def make_poverty_classifier(threshold: float):
    """Return a function classifying income against an absolute threshold."""
    def classify(equivalised_income: float) -> str:
        return "at risk" if equivalised_income < threshold else "not at risk"
    return classify
````
