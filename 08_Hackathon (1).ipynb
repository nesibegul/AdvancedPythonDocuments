````python
# Helper: embed a screenshot in your notebook as a base64 image.
# Handy when your demo is slow, external, or can't run live —
# the image travels *inside* the .ipynb, so nothing breaks when you email it.
import base64
from pathlib import Path
from IPython.display import HTML, display


def embed_image(path: str | Path, width: int | None = None) -> str:
    """Display a local image inline, base64-encoded into the notebook."""
    raw = Path(path).read_bytes()
    b64 = base64.b64encode(raw).decode("ascii")
    suffix = Path(path).suffix.lstrip(".").lower()
    mime = "jpeg" if suffix in ("jpg", "jpeg") else suffix
    size = f' width="{width}"' if width else ""
    html = f'<img src="data:image/{mime};base64,{b64}"{size}>'
    display(HTML(html))
    return html

NOTEBOOK_DIR = Path.cwd()
FIGURES = NOTEBOOK_DIR.parent.parent / "figures"

# Example — point it at your own screenshot:
embedding = embed_image(FIGURES / "silc_choropleth.png", width=720)
print("embed_image() ready — drop your screenshots straight into the notebook for the talk")
print(embedding[:100])
````

## 7 · Coaching Tips

### When You Are Stuck

1. **Read the error message** — Python errors are usually very specific
2. **Check the data types** — `df.dtypes`, `type(x)`, `print(x[:5])`
3. **Simplify** — test your function on 10 rows before running on 100,000
4. **Ask a teammate** — pair programming catches bugs faster than solo debugging
5. **Ask the trainer** — that's what I'm here for!

### What Makes a Great Demo

- **It runs live** — pre-recorded or screenshots are fine, but live is most impressive
- **Real data** — use actual numbers from your problem, not toy examples
- **Story** — *"We wanted to answer X; here's how Python helped"*
- **Edge cases handled** — show what happens with missing or malformed data
- **Code quality** — tests, type hints, logging (your reviewers are programmers too)

### Common Pitfalls to Avoid

- **Over-scoping** — a working demo of one feature beats five broken ones
- **Skipping tests** — 3 good pytest tests take 15 minutes and save hours of debugging
- **No error handling** — always guard external I/O with try/except
- **Hardcoded paths** — use `Path(__file__).parent / "data"`, never `C:\Users\you\...`
